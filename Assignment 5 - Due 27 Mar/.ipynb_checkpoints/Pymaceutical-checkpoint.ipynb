### Summary of Key Findings

* Treatment with capomulin led to an 84% survival rate after 45 days compared to 44-52% survival for placebo or either of infubinol or ketrapil
* Of capomulin, infubinol, and ketrapil, only treatment with capomulin led to a statistically significant (at 95% confidence) decrease in tumor volumes 
* Statistically significant positive outcomes were observed for treatment with capomulin both in terms of a reduction in tumor volumes and a reduction in metastatic sites.  

```python
# Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')

# File to Load (Remember to Change These)
mouse_data = 'data/mouse_drug_data.csv'
clinical_data = 'data/clinicaltrial_data.csv'

# Read the Mouse and Drug Data and the Clinical Trial Data

mouse_df = pd.read_csv(mouse_data)
clin_df = pd.read_csv(clinical_data)
# mouse_df maps mouse ID to drug
# clin_df maps mouse ID to timepoint, tumor volume, and metastatic sites
# survival rates will need to be derived from the number of mice for which there is data

# Combine the data into a single dataset
data_df = clin_df.merge(mouse_df, on='Mouse ID', how = 'outer')

# Display the data table for preview
data_df.head()
```

## Tumor Response to Treatment

```python
# Store the Mean and Standard Error of Tumor Volume for analysis in four separate DataFrames
# Separate DataFrames will enable us to plot more easily than one big frame to be picked apart later
# It is also easier to debug if the identical steps are grouped by fours
tvol_mean = data_df[['Drug','Timepoint','Tumor Volume (mm3)']].groupby(['Drug','Timepoint']).mean().reset_index()
tvol_sem = data_df[['Drug','Timepoint','Tumor Volume (mm3)']].groupby(['Drug','Timepoint']).sem().reset_index()
# Do the same for Metastatic Sites
ms_mean = data_df[['Drug','Timepoint','Metastatic Sites']].groupby(['Drug','Timepoint']).mean().reset_index()
ms_sem = data_df[['Drug','Timepoint','Metastatic Sites']].groupby(['Drug','Timepoint']).sem().reset_index()
# Print one DataFrame for later reference
tvol_mean.head(10)
```

```python
# Now use a pivot to reformat the four DataFrames for plotting.  Reset the index so that for 'scatter'
# plot type, the times are available as a column ... 
tvol_mean_bydrug = tvol_mean.pivot(index='Timepoint',columns='Drug').reset_index()
tvol_sem_bydrug = tvol_sem.pivot(index='Timepoint',columns='Drug').reset_index()
ms_mean_bydrug = ms_mean.pivot(index='Timepoint',columns='Drug').reset_index()
ms_sem_bydrug = ms_sem.pivot(index='Timepoint',columns='Drug').reset_index()
# To make it easier to handle, the column names will be re-set to just the drug names
tvol_mean_bydrug.columns = tvol_mean_bydrug.columns.get_level_values(1)
tvol_sem_bydrug.columns = tvol_sem_bydrug.columns.get_level_values(1)
ms_mean_bydrug.columns = ms_mean_bydrug.columns.get_level_values(1)
ms_sem_bydrug.columns = ms_sem_bydrug.columns.get_level_values(1)
# Check that the DataFrames are plot-ready
tvol_mean_bydrug.head()
```

```python
# Generate the Plot (with Error Bars) for each drug. 
# Tumor volume first, then will repeat with metastatic sites.
# Since these are multiple scatter plots on the same axis, 
# an axis set will be generated first and then called as reference.  
# Plot only Capomulin, Infubinol, Ketapril, and Placebo per instructions
ax1 = tvol_mean_bydrug.plot(kind='scatter', x = 0, y = 'Capomulin', yerr = tvol_sem_bydrug['Capomulin'],
                            color = 'darkcyan', label = 'Capomulin')
tvol_mean_bydrug.plot(kind='scatter', x = 0, y = 'Infubinol', yerr = tvol_sem_bydrug['Infubinol'],
                      color = 'darkblue', label = 'Infubinol', ax = ax1)
tvol_mean_bydrug.plot(kind='scatter', x = 0, y = 'Ketapril', yerr = tvol_sem_bydrug['Ketapril'], 
                      color = 'darkviolet', label = 'Ketapril',  ax = ax1)
tvol_mean_bydrug.plot(kind='scatter', x = 0, y = 'Placebo', yerr = tvol_sem_bydrug['Placebo'],
                      color = 'black', label = 'Placebo', ax = ax1)
ax1.set_xlabel('Time (days)')
ax1.set_ylabel('Tumor Volume (mm$^3$)')
ax1.set_xlim(0, 50)
ax1.set_ylim(30, 80)
ax1.legend(loc='best')
ax1.set_title('Treatment with Capomulin Reduced Tumor Volume')
plt.savefig('Output_Figures/tumor_volume.jpg')
```

```python
# Show the Figure
plt.show()
```

## Metastatic Response to Treatment

```python
# Generate the Plot (with Error Bars) for each drug. 
# Now metastatic sites ... 
# Since these are multiple scatter plots on the same axis, 
# an axis set will be generated first and then called as reference.  
# Plot only Capomulin, Infubinol, Ketapril, and Placebo per instructions
ax1 = ms_mean_bydrug.plot(kind='scatter', x = 0, y = 'Capomulin', yerr = ms_sem_bydrug['Capomulin'],
                            color = 'darkcyan', label = 'Capomulin')
ms_mean_bydrug.plot(kind='scatter', x = 0, y = 'Infubinol', yerr = ms_sem_bydrug['Infubinol'],
                      color = 'darkblue', label = 'Infubinol', ax = ax1)
ms_mean_bydrug.plot(kind='scatter', x = 0, y = 'Ketapril', yerr = ms_sem_bydrug['Ketapril'], 
                      color = 'darkviolet', label = 'Ketapril',  ax = ax1)
ms_mean_bydrug.plot(kind='scatter', x = 0, y = 'Placebo', yerr = ms_sem_bydrug['Placebo'],
                      color = 'black', label = 'Placebo', ax = ax1)
ax1.set_xlabel('Time (days)')
ax1.set_ylabel('Metastatic Sites')
ax1.set_xlim(0, 50)
ax1.set_ylim(0, 4)
ax1.legend(loc='best')
ax1.set_title('Treatment with Capomulin Yielded Fewer Metastatic Sites')
plt.savefig('Output_Figures/metastatic_sites.jpg')
```

## Survival Rates

```python
# Generate a new DataFrame with survival rates based on mouse ID unique counts 
# We could infer a standard error if we made some assumptions, but we will refrain
# First, generate the counts
surv = data_df[['Drug','Timepoint','Mouse ID']].groupby(['Drug','Timepoint'])[['Mouse ID']].nunique().reset_index()
# Then pivot this DataFrame as before
surv_bydrug = surv.pivot(index='Timepoint',columns='Drug')
# Simplifh column labels
surv_bydrug.columns = surv_bydrug.columns.get_level_values(1)
# Convert to survival rate by dividing by initial values
for col in surv_bydrug:
    surv_bydrug[col] = surv_bydrug[col] / surv_bydrug.at[0, col] * 100
# Reset index so that we have a column for scatterplot
surv_bydrug = surv_bydrug.reset_index()
# Check that everything worked as intended
surv_bydrug.head(10)
```

```python
# Now let's make a scatterplot as before 
# Since these are multiple scatter plots on the same axis, 
# an axis set will be generated first and then called as reference.  
# Plot only Capomulin, Infubinol, Ketapril, and Placebo per instructions
ax1 = surv_bydrug.plot(kind='scatter', x = 0, y = 'Capomulin',
                            color = 'darkcyan', label = 'Capomulin')
surv_bydrug.plot(kind='scatter', x = 0, y = 'Infubinol',
                      color = 'darkblue', label = 'Infubinol', ax = ax1)
surv_bydrug.plot(kind='scatter', x = 0, y = 'Ketapril', 
                      color = 'darkviolet', label = 'Ketapril',  ax = ax1)
surv_bydrug.plot(kind='scatter', x = 0, y = 'Placebo',
                      color = 'black', label = 'Placebo', ax = ax1)
ax1.set_xlabel('Time (days)')
ax1.set_ylabel('Survival (%)')
ax1.set_xlim(0, 50)
ax1.set_ylim(0, 100)
ax1.legend(loc='best')
ax1.set_title('Treatment with Capomulin Improved Survival')
plt.savefig('Output_Figures/survival_rates.jpg')
```

## Summary Bar Graph

```python
# Calculate the percent changes for each drug, and the standard error in each
shrink_bydrug = {}
shrinkerr_bydrug = {}
# Build a dictionary so we can select by drug easily later 
# Ignore the first column of timepoints, which we had reserved in order to make scatterplots
for col in tvol_mean_bydrug.iloc[:,1:]:
    # Use the first and last row values in each column to calculate shrinkage to make the method general
    shrink_bydrug.update({col:( - 100.0 * (tvol_mean_bydrug.at[len(tvol_mean_bydrug) - 1, col] / tvol_mean_bydrug.at[0, col] -1.0))})
    # Compute the error by normalizing the standard error of volume change (assumes initial value is known perfectly)
    shrinkerr_bydrug.update({col:( - 100.0 * (tvol_sem_bydrug.at[len(tvol_sem_bydrug) - 1, col] / tvol_mean_bydrug.at[0, col]))})
    # Display the data to confirm
shrink_bydrug, shrinkerr_bydrug
```

```python
# Next, we'll build plot objects for the drugs that we want.  
# This will make it easy to build a summary using any of the data we want.  
Selected_Drugs = ['Capomulin','Infubinol','Ketapril','Placebo']
# To build a bar chart, we need x-axis values, y-values, error bars, labels, colors, and tick marks
# We can conveniently use the Selected_Drugs list as our "x" object, labels, and tick marks all in one!
x_values = Selected_Drugs
# Initialize the y-values, error values, and colors as empty lists
y_values = []
error_values = []
colorlist = []
# Now loop through all the drugs and all the ones we want to each list
for drug_name in Selected_Drugs:
    y_values.append(shrink_bydrug[drug_name])
    error_values.append(shrinkerr_bydrug[drug_name])
    # Set the color based on tumor shrinkage.  If positive, color green, otherwise color red 
    if shrink_bydrug[drug_name] > 0:
        colorlist.append('g')
    else:
        colorlist.append('r')
        
# Check our lists
x_values, y_values, error_values, colorlist
```

```python
# Finally, make the plot
#Set figure size
plt.figure(figsize = (6,4))
# Make bar chart and save axes item for later
ax = plt.bar(x_values,y_values, width = 0.75, yerr = error_values, color = colorlist, alpha = 0.6)
# Draw line, add titles, set limits
plt.hlines(0, -0.5, 3.5)
plt.xlabel('Drug Candidate')
plt.ylabel('Reduction in Tumor Volume After 45 days (%)')
plt.title('Promising Reults from Studies of Capomulin for Treating SCC')
plt.ylim(-70, 50)
# Generate the bar coordinates for placing labels
rects = ax.patches
# Generate the label text items (round to nearest %)
bar_labels = [f'{val:.0f}' for val in y_values]
# Standard 'add text' method from Stack Overflow, adjusted for text placement
for rect, label in zip(rects, bar_labels):
    height = rect.get_height()
    plt.text(rect.get_x() + rect.get_width() / 2, height + 7 * np.sign(height), label,
             ha='center', va='center', size = 12)

plt.savefig('Output_Figures/summary.jpg')
```
